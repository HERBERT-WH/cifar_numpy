```python
#对数据进行分析
%matplotlib inline
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
meta_path = "./cifar-10-batches-py/batches.meta"
with open(meta_path, 'rb') as f:
    meta_data = pickle.load(f, encoding='latin1')
    print(meta_data)
```

```python
def cifar_data(data_dir):
    with open(data_dir, 'rb') as f:
        data = pickle.load(f, encoding='latin1')
        X = data['data']
        Y = data['labels']
        X = X.reshape((-1, 3, 32, 32))
        X = np.transpose(X, [0, 2, 3, 1])
        Y = np.array(Y)
    return X, Y
```

```python
data_path = "./cifar-10-batches-py/data_batch_1"
X, Y = cifar_data(data_path)
print(X.shape) #每个batchzhose是10000张图片，每个图片是32*32*3
```

```python
# 显示示例图片，查看数据
show_number = 10

# 设置图像大小和布局
plt.figure(figsize=(20, 5))  # 设置整体图像大小，宽度20，高度2
plt.subplots_adjust(wspace=0.3)  # 调整子图之间的间距

for i in range(show_number):
    img = X[i]
    plt.subplot(1, show_number, i + 1)
    plt.imshow(img.astype('uint8'))
    plt.title(f'lable:{Y[i]}') 
    plt.axis('off')
```
